```python
import requests
import pandas as pd
from csv import reader
from bs4 import BeautifulSoup
read_obj = pd.read_csv(r'C:\Users\Tianx\Desktop\HIWI\Task2\final_df_outage.csv')
#print(read_obj)
#read_obj['postalCode'][0]
#read_obj=read_obj.loc[:10]
base_url='https://www.europacco.com/en/find-zip/de?country=de&search_string='
i=1
def add_zero(x):
    y=str(x)
    while(len(y)<5):
        y='0'+y
    return y  
state=[]
postalcode=[]
for x in read_obj['postalCode']:
    page = requests.get(base_url+add_zero(x)) 
    #print(x)
    #print(page)
    html=page.content
    soup=BeautifulSoup(html,'html.parser')
    td_list = soup.find_all('td')
    if(td_list == []):
        i=i+1
        print('****************None',i)
        state.append('none')
        postalcode.append(add_zero(x))
    else:
        #print(td_list)
        print(td_list[1].text,i)
        i=i+1
        #print(i)
        state.append(td_list[1].text)
        postalcode.append(add_zero(x))
    #print(state)
_data = pd.DataFrame()
_data['postalcode']=postalcode
_data['state']=state
_data.to_csv('task2_t2.csv' , index=False,sep=',', encoding='utf-8')
```
